```python

def index_creator(stock_industry):
    #Random Ticker to get column data and assign zeroes to the index dataframe
    stock_index = pdr.get_data_yahoo('MSFT', start="2020-03-20", end="2021-07-07") 
    for col in stock_index.columns:
        stock_index[col].values[:] = 0

    for row in stock_industry:
        current_df = pdr.get_data_yahoo(row, start="2020-03-20", end="2021-07-07")
        current_df = current_df.div(10)
        stock_index['Open'] += current_df['Open']
        stock_index['High'] += current_df['High']
        stock_index['Low'] += current_df['Low']
        stock_index['Close'] += current_df['Close']
        stock_index['Adj Close'] += current_df['Adj Close']
        stock_index['Volume'] += current_df['Volume']
    return stock_index

basic_industry_stocks_index = index_creator(basic_industry_stocks)
capital_goods_stocks_index = index_creator(capital_goods_stocks)
consumer_discretionary_stocks_index = index_creator(consumer_discretionary_stocks)
energy_stocks_index = index_creator(energy_stocks)
consumer_services_stocks_index = index_creator(consumer_services_stocks)
finance_stocks_index = index_creator(finance_stocks)
healthcare_stocks_index = index_creator(healthcare_stocks)
public_utilities_stocks_index = index_creator(public_utilities_stocks)
real_estate_stocks_index = index_creator(real_estate_stocks)
technology_stocks_index = index_creator(technology_stocks)
transportation_stocks_index = index_creator(transportation_stocks)

```

```python
import numpy as np
from pandas_datareader import data as pdr
datasetDf     = pdr.get_data_yahoo('MSFT', start="2020-03-20", end="2021-07-07")

# Create vertical array of features.
ROW_DIM       = 0
COL_DIM       = 1
openPrice     = datasetDf[['Open']].values
verticalArray = openPrice.reshape(openPrice.shape[ROW_DIM], 1)

# Scale open prices.
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler(feature_range = (0, 1))
openPriceScaled = sc.fit_transform(verticalArray)

# Compare data with transformations.
print("\nDataFrame")
print(datasetDf.head(3))
print(datasetDf.tail(3))

print("\nOpen Price")
print(openPrice)

print("\nOpen Price Scaled")
print(openPriceScaled)

# Creating a data structure with 60 prior time steps and 1 current output.
X = []
y = []
NUM_TIME_STEPS = 60
for i in range(NUM_TIME_STEPS, len(openPriceScaled)):
    X.append(openPriceScaled[i-NUM_TIME_STEPS:i, 0])

for i in range(0, len(openPriceScaled)-NUM_TIME_STEPS):
    if(i==0):
        y.append(np.NaN)
    else:
        y.append(openPriceScaled[i-1][0])

X, y = np.array(X), np.array(y)
X    = X.reshape(X.shape[0], X.shape[1], 1)

print("\nOpen Price Scaled")
print(openPriceScaled)
print("\nX-Features")
print(X)
print("\nY-values")
print(y)

# Split into test and training sets #######
TEST_SET_SiZE = 30
X_test  = X[:TEST_SET_SiZE]
y_test  = y[:TEST_SET_SiZE]

# Remove last day since y is unknown.
y_test_t = y_test[1:len(y_test)-1,]
X_test_t = X_test[1:len(X_test)-1,]

# Store X for next day prediction.
X_current = X_test[0,]

X_train  = X[TEST_SET_SiZE:]
y_train  = y[TEST_SET_SiZE:]

# Part 2 - Building the RNN

# Importing the Keras libraries and packages
plt.style.use('seaborn')
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout

def create_model(X_train, y_train, X_test, y_test):
    # Initialize the RNN.
    regressor = Sequential()

    # Adding the first LSTM layer and some Dropout regularisation
    regressor.add(LSTM(units = 50, return_sequences = True,
                       input_shape = ( X_train.shape[1],1)))
    regressor.add(Dropout(0.2))

    # Adding a second LSTM layer and some Dropout regularisation
    regressor.add(LSTM(units = 50, return_sequences = True))
    regressor.add(Dropout(0.2))

    # Adding a third LSTM layer and some Dropout regularisation
    regressor.add(LSTM(units = 50, return_sequences = True))
    regressor.add(Dropout(0.2))

    # Adding a fourth LSTM layer and some Dropout regularisation
    regressor.add(LSTM(units = 50))
    regressor.add(Dropout(0.2))

    # Adding the output layer
    regressor.add(Dense(units = 1))

    # Compiling the RNN.
    regressor.compile(optimizer = 'adam', loss = 'mean_squared_error',
                      metrics=['accuracy'])

    # Fit model.
    history = regressor.fit(X_train, y_train, epochs = 100, batch_size = 32,
                            validation_data=(X_test, y_test))
    return regressor, history

regressor, history  = create_model(X_train, y_train, X_test_t, y_test_t)

# Make predictions and de-scale so prices are realistic.
raw_predictions      = regressor.predict(X_test_t)
predicted_prices     = sc.inverse_transform(raw_predictions)

# Reshape y back to manageable format.
# Convert back to real unscaled prices.
actualPricesReshaped = y_test_t.reshape(len(y_test_t),1)
real_stock_prices    = sc.inverse_transform(actualPricesReshaped)

# Reverse sequence for graphing so latest date appears last.
def reverseOrder(list):
    reversedList = []
    lengthList = len(list)
    for i in range(lengthList-1, -1, -1):
        reversedList.append(list[i])
    return reversedList

# Visualising the predicted versus actual.
def showPredictions(predicted, actual):
    plt.plot(actual, color = 'green',
             label = 'Real Microsoft Stock Price')
    plt.plot(predicted, color = 'blue',
             label = 'Predicted Microsoft Stock Price')
    plt.title('Microsoft Stock Price Prediction')
    plt.xlabel('Time')
    plt.ylabel('Microsoft Stock Price')
    plt.legend()
    plt.show()

# Reverse order so most current day  appears last.
predicted = reverseOrder(predicted_prices)
actual    = reverseOrder(real_stock_prices)
showPredictions(predicted, actual)

def showLoss(history):
    # Get training and test loss histories
    training_loss       = history.history['loss']
    validation_loss     = history.history['val_loss']

    # Create count of the number of epochs
    epoch_count = range(1, len(training_loss) + 1)

    # Visualize loss history for training data.
    plt.plot(epoch_count, training_loss, label='Train Loss', color='red')

    # View loss on unseen data.
    plt.plot(epoch_count, validation_loss, 'r--', label='Validation Loss',
             color='black')
  
    plt.xlabel('Epoch')
    plt.legend(loc="best")
    plt.title("Loss")
    plt.show()

showLoss(history)

# Show RMSE.
from sklearn import metrics
print('Root Mean Squared Error:',
      np.sqrt(metrics.mean_squared_error(real_stock_prices, predicted_prices)))


```
